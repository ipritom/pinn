```python
def get_pos(v, theta, t):
    theta = np.deg2rad(theta)
    x = v*np.cos(theta)*t
    y = -0.5*9.8*t**2 + v*np.sin(theta)*t

    return x, y
```

```python
import math

def projectile_position(v0, theta_degrees, t):
    # Convert angle to radians
    theta_radians = math.radians(theta_degrees)
    
    # Calculate initial velocity components
    v0x = v0 * math.cos(theta_radians)
    v0y = v0 * math.sin(theta_radians)
    
    # Calculate x and y positions
    g = 9.8  # acceleration due to gravity in m/s^2
    x = v0x * t
    y = -0.5 * g * t**2 + v0y * t
    
    T = 2*(v0*math.cos(theta_radians)/g)
    
    return x, y

# Example usage
# v0 = 35  # initial velocity in m/s
# theta = 45  # launch angle in degrees
# t = 2  # time in seconds

# x, y = projectile_position(v0, theta, t)
# print(f"At t = {t} s:")
# print(f"x position: {x:.2f} m")
# print(f"y position: {y:.2f} m")
```

```python
def get_data(duration, step, v0, theta):
    X = []
    Y = []
    for t in np.arange(0,duration,step=step):
        x, y = projectile_position(v0, theta, t)
        if t==0:
            X.append(x)
            Y.append(y)
            continue

        if y<=0:
            break
        X.append(x)
        Y.append(y)
    
    return X, Y 
```
